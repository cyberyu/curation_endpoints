# Pretrained  Classification Models

```python
api_select=widgets.Dropdown(
    options=['finbert_sentiment','movie_review_sentiment','zero_shot_classification'],
    value='finbert_sentiment',
    description='Classification model:',
    disabled=False
)

text= widgets.Textarea(
    value=' '.join(['Stocks rallied and the British pound gained.', "There wasn't much change in the market today.", "Telsa's stock price plummeted."]),
    placeholder='Type something',
    description='Text:',
    disabled=False, 
    layout=l
)

labels=widgets.Textarea(
    value='',
    placeholder='Type something',
    description='Labels:',
    disabled=False, 
    
)

url_list={'finbert_sentiment':'https://domino4.domino.opsp.c1.vanguard.com:443/models/61437673a385bc49679f8752/latest/model',
          'movie_review_sentiment':'https://domino4.domino.opsp.c1.vanguard.com:443/models/614378f1a385bc49679f8789/latest/model',
          'zero_shot_classification':'https://domino4.domino.opsp.c1.vanguard.com:443/models/61436b0612d29b6ae658644e/latest/model'
          }
   
    
apikey_list={'finbert_sentiment':'0jMOoiSfzdhuRFlNAPHEfUxStJdg9Lkl1pmo4UozzS9ZgNZlhUzSSIeuKmSFtTrZ',
          'movie_review_sentiment':'p98GviYdeSHCYS1p6skpLp2AntlqiBl18aWOCRA2RH6rdkx0IdF3WsrTmIibCeQO',
          'zero_shot_classification':'cZJjphvoE2vjt6MqCoGF0kU9rJIvI77SXjS6aOZyy42jHLGjrCx7pPTAjhjraloR'}    


l = Layout(flex='0 1 auto', height='40px', min_height='40px', width='auto')

button=widgets.Button(
    description='submit query',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='submit query',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)

def on_button_clicked(b):
    with output:
        sentences = text.value.split(",")
        label_names = ['travel', 'cooking', 'dancing']  # NOTE: can choose any label names we want.  
        
        if (api_select.value=='finbert_sentiment') or (api_select.value=='movie_review_sentiment'):
            label_names=None 
        elif api_select.value=='zero_shot_classification':
            label_names=['travel', 'cooking', 'dancing'] 
            
        response = requests.post(url_list[api_select.value],
            auth=(apikey_list[api_select.value],apikey_list[api_select.value]),
            json={
                "data": {"sentences": sentences, "label_names": label_names}
            },
            verify=False
        )        
                       
        print(response.json())
        
        
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        if change['new']=='finbert_sentiment':
            text.value=', '.join(['Stocks rallied and the British pound gained.', "There wasn't much change in the market today.", "Telsa's stock price plummeted."])
            labels.value=''
        elif change['new']=='movie_review_sentiment':
            text.value=', '.join(['I really liked that.', 'Why was it so boring.',  "There wasn't much change in the market today.", "Telsa's stock price plummeted."])
            labels.value=''
        elif change['new']=='zero_shot_classification':
            text.value=', '.join(['one day I will see the world.', 'I love making pasta with my grandma.', "I'm making pasta today."])
            labels.value='travel, cooking, dancing'
        with output:
            print("changed to %s" % change['new'])
        
button.on_click(on_button_clicked) 
api_select.observe(on_change)

output = widgets.Output()

ui = widgets.VBox([widgets.HBox([api_select, labels]), text, button, output])

ui
```
